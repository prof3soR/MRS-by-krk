importing all the libraries and the csv file of netflix data


```python
import pandas as pd
import numpy as np
import difflib
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer 
netflix=pd.read_csv('movies.csv')
```

Printing the head of the csv file

```python
features=['genres','keywords','tagline','cast','director','overview']
```

```python
for i in features:
    netflix[i]=netflix[i].fillna('')
```

Combning all the similar fearures

```python
all_features=netflix['cast']+' '+netflix['genres']+' '+' '+netflix['keywords']+' '+netflix['tagline']+netflix['director']+netflix['overview']
```

Converting the text data into feature vectors

```python
vect=TfidfVectorizer()
```

```python
feature_vector=vect.fit_transform(all_features)
```

Getting the similarity score using cosine similarity 

```python
similarity = cosine_similarity(feature_vector)
```

creating all the movie names given in the data set (netflix)

```python
all_titles=netflix['title'].tolist()
```

Fininda the close the match for the movie name given by the user

```python
movie=input("Enter your movie name : ")
```

```python
all_close_match=difflib.get_close_matches(movie,all_titles)
print(all_close_match)
```

```python
close_match=all_close_match[0]
print(close_match)
```

```python
index=netflix[netflix.title==close_match]['index'].values[0]
print(index)
```

```python
similarity_score=list(enumerate(similarity [index]))
print(similarity_score)
```

```python
sorted_movies=sorted(similarity_score, key=lambda x:x[1],reverse=True)
print(sorted_movies)
```

```python
print("Moive you might like : ")
count=1
for i in sorted_movies:
    index=i[0]
    title_from_data=netflix[netflix.index==index]['title'].values[0]
    if count<10:
        print(title_from_data)
        count+=1
    
    
```

Final code for output :-)

```python
movie=input("Enter your movie name : ")
all_close_match=difflib.get_close_matches(movie,all_titles)
close_match=all_close_match[0]
index=netflix[netflix.title==close_match]['index'].values[0]
similarity_score=list(enumerate(similarity [index]))
sorted_movies=sorted(similarity_score, key=lambda x:x[1],reverse=True)
print("Moives you might like : ")
count=1
for i in sorted_movies:
    index=i[0]
    title_from_data=netflix[netflix.index==index]['title'].values[0]
    if count<=10:
        print(str(count)+'. '+title_from_data)
        count+=1
    
```
